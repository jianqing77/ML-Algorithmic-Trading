### Using **K-Nearest Neighbors (KNN)** algorithm for a **classification task**

#### Sample 1

````python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
````

````python
# Sample dataset

# Input Data -- Feature Matrix
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]]) # 4 data points, each having 2 features (2D space)

# Output Labels -- Target Vector
y = np.array([0, 0, 1, 1])

# Model - creating the KNN classifier
# n_neighbors specifies the number of nearest neighbors to consider when making a prediction. 
# In this case, for any new data point, the model will look at the 3 closest points in the training data (X) to decide its class
neigh = KNeighborsClassifier(n_neighbors=3) 

# Fitting the model: trains the KNN model using the feature matrix X and the corresponding labels y
neigh.fit(X, y)

# Predicting a new data point - use majority voting
print(neigh.predict([[2.5, 2.5]]))
````

In this example, we are using the **K-Nearest Neighbors (KNN)** algorithm for a **classification task**. The steps outlined in our code demonstrate how to create a KNN model, fit it to some data, and use it to make a prediction on a new data point.

Let’s break it down step by step.

1. **Sample Dataset**

```python
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
y = np.array([0, 0, 1, 1])
```

- **`X`**: This is your **feature matrix** (input data) with 4 data points, each having 2 features (2D space).
  - `[1, 1]`, `[1, 2]`, `[2, 2]`, `[2, 3]` represent the **coordinates** of these points in a 2D space.
  
- **`y`**: This is your **target vector** (output labels), indicating the class labels for each corresponding data point in `X`.
  - `[0, 0, 1, 1]` means that:
    - The first two points (`[1, 1]` and `[1, 2]`) belong to **class 0**.
    - The last two points (`[2, 2]` and `[2, 3]`) belong to **class 1**.
  
2. **Creating the KNN Classifier**

```python
neigh = KNeighborsClassifier(n_neighbors=3)
```

- **`KNeighborsClassifier`**: This is the K-Nearest Neighbors classifier from the **scikit-learn** library.
- **`n_neighbors=3`**: This specifies the number of **nearest neighbors** to consider when making a prediction. In this case, for any new data point, the model will look at the **3 closest points** in the training data (`X`) to decide its class.

3. **Fitting the Model**

```python
neigh.fit(X, y)
```

- **`fit(X, y)`**: This method **trains the KNN model** using the feature matrix `X` and the corresponding labels `y`. 
  - The model will store the examples in `X` and use them later to make predictions based on the distances between new data points and the existing points in `X`.

KNN is a **lazy learner**, meaning it doesn’t actually "learn" a model in the traditional sense during training but rather **memorizes** the training data. Predictions are made directly based on the stored data points.

4. **Predicting a New Data Point**

```python
print(neigh.predict([[2.5, 2.5]]))
```

- **`neigh.predict([[2.5, 2.5]])`**: This line uses the trained KNN model to predict the class label for a **new data point** `[2.5, 2.5]`.
  - The model will find the **3 nearest neighbors** to `[2.5, 2.5]` based on the **Euclidean distance** (by default) and then assign the class label based on the **majority class** of those neighbors.

Let’s calculate what happens during this prediction step:

**Step 1: Calculate the Euclidean Distances**

The model will calculate the **Euclidean distance** between the new point `[2.5, 2.5]` and each of the four points in the training set `X`:

- Distance to `[1, 1]`:  
  \[\sqrt{(2.5 - 1)^2 + (2.5 - 1)^2} = \sqrt{1.5^2 + 1.5^2} = \sqrt{2.25 + 2.25} = \sqrt{4.5} \approx 2.12 \]
- Distance to `[1, 2]`:  
  \sqrt{(2.5 - 1)^2 + (2.5 - 2)^2} = \sqrt{1.5^2 + 0.5^2} = \sqrt{2.25 + 0.25} = \sqrt{2.5} \approx 1.58
- Distance to `[2, 2]`:  
  \sqrt{(2.5 - 2)^2 + (2.5 - 2)^2} = \sqrt{0.5^2 + 0.5^2} = \sqrt{0.25 + 0.25} = \sqrt{0.5} \approx 0.71
- Distance to `[2, 3]`:  
  \[\sqrt{(2.5 - 2)^2 + (2.5 - 3)^2} = \sqrt{0.5^2 + (-0.5)^2} = \sqrt{0.25 + 0.25} = \sqrt{0.5} \approx 0.71 \]

**Step 2: Find the 3 Nearest Neighbors**

The **3 nearest neighbors** to the new point `[2.5, 2.5]` are:
1. `[2, 2]` with a distance of **0.71** (Class 1)
2. `[2, 3]` with a distance of **0.71** (Class 1)
3. `[1, 2]` with a distance of **1.58** (Class 0)

**Step 3: Majority Voting**

The model looks at the **classes of the 3 nearest neighbors**:
- Two neighbors belong to **Class 1**.
- One neighbor belongs to **Class 0**.

Since Class 1 is the majority, the model predicts that the new point `[2.5, 2.5]` belongs to **Class 1**.

**Output:**

```
[1]
```

This means the model predicts that the new data point `[2.5, 2.5]` belongs to **Class 1**.

**Summary of Key Steps:**

1. **KNN Model Creation**:
   - You create a KNN classifier with `n_neighbors=3`, meaning it will consider the 3 nearest neighbors for making predictions.

2. **Model Training**:
   - The model is trained using the dataset (`X`, `y`), where `X` contains the feature points and `y` contains the corresponding class labels.

3. **Making Predictions**:
   - For a new data point `[2.5, 2.5]`, the model finds the 3 nearest neighbors, checks their class labels, and assigns the majority class as the predicted label.
   - In this case, the new point is classified as **Class 1**.

**Conclusion:**

K-Nearest Neighbors (KNN) is a simple, intuitive algorithm that classifies data points based on the majority class of their nearest neighbors in the training set. In this example, the new point `[2.5, 2.5]` was classified as **Class 1** because the majority of its 3 closest neighbors belong to Class 1.

#### Sample 2

````python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
````

````python
# Load sample data (Iris dataset)
iris = datasets.load_iris()
X = iris.data[:, :2]  # Using only the first two features for simplicity
y = iris.target

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the data into training and test sets - random
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
knn.fit(X_train, y_train)
````

````python
# Plotting
x_min, x_max = X_scaled[:, 0].min() - 1, X_scaled[:, 0].max() + 1
y_min, y_max = X_scaled[:, 1].min() - 1, X_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=y, s=20, edgecolor='k')
plt.title('KNN (k=5) with Iris Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
````
